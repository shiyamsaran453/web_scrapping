```python
from bs4 import BeautifulSoup as bs
```

```python
import pandas as pd
```

```python
import requests as re
```

```python
url="https://www.flipkart.com/sports/cricket/cricket-bat/english-willow~willow-type/pr?sid=abc%2C5lf%2Cp8v&p%5B%5D=facets.age_group%255B%255D%3D15%252B%2BYrs&p%5B%5D=facets.rating%255B%255D%3D4%25E2%2598%2585%2B%2526%2Babove&p%5B%5D=facets.willow_type%255B%255D%3DEnglish%2BWillow"
```

```python
req=re.get(url)
```

```python
req
```

```python
back_code=bs(req.text)
```

```python
back_code
```

```python
bat_name=back_code.select(".s1Q9rs")
```

```python
ratings=back_code.select("._3LWZlK")
```

```python
amount_with_discount=back_code.select("._25b18c")
```

```python
delivery=back_code.select("._2hu4Aw ._2Tpdn3")
```

```python
bat_data=pd.DataFrame(bat_name,columns=["BAT_NAME"])
```

```python
bat_data["RATINGS"]=ratings
```

```python
bat_data["AmountWithDiscount"]=amount_with_discount
```

```python
bat_data["Delivery"]=delivery
```

```python
bat_data.to_csv("b.csv")
```
